```python
import yfinance as yf
import pandas as pd
import pandas_ta as pta
import matplotlib.pyplot as plt
```

```python
msft =yf.Ticker("MSFT").history(period="14d",interval = "1h")
y=msft['Close'].tolist()
x=msft.index.values
y
```

```python
rsi =pta.rsi(msft['Close'], length = 14)
rsi.fillna(0, inplace=True)
z=rsi.tolist()
z
```

```python
plt.figure(figsize=(20,15)) 

plt.plot(x, y, color='r', label='Close')
plt.plot(x, z, color='g', label='rsi')
  
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("Date")
plt.title("RSI")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
  
# To load the display window
plt.show()
```
